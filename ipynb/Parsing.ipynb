```python
import numpy as np
import pandas as pd
import spacy
nlp = spacy.load("en_core_web_sm")
from spacy import displacy
```

```python
data_file = '../data/with_narrative.csv'
df = pd.read_csv(data_file)
```

```python
text = 'XXXX switched to XXXX i found out after they were gone after in 20 states by federal gov. and settlements was done which I was not aware of.'
doc = nlp(text)
```

```python
displacy.serve(doc)
```

```python
for token in doc:
    print(token.text, token.lemma_, token.pos_, token.tag_, token.dep_,
            token.shape_, token.is_alpha, token.is_stop)
```
